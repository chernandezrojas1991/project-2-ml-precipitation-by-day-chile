**Obtener mes y estación del año según fecha y eliminar columna fecha**

Crear función para determinar estación del año en Chile según fecha

```python
def obtener_estacion_chile(fecha):
    mes = fecha.month
    dia = fecha.day
    if (mes == 12 and dia >= 21) or (mes in [1, 2]) or (mes == 3 and dia <= 20):
        return 'Verano'
    elif (mes == 3 and dia >= 21) or (mes in [4, 5]) or (mes == 6 and dia <= 20):
        return 'Otoño'
    elif (mes == 6 and dia >= 21) or (mes in [7, 8]) or (mes == 9 and dia <= 20):
        return 'Invierno'
    else:
        return 'Primavera'
```

Convertir columna fecha en tipo datetime

```python
df_precip_clean['fecha'] = pd.to_datetime(df_precip_clean['fecha'])
```

Crear columnas 'mes' y 'estación' extraídas y procesadas a través de la columna 'fecha', la cual corresponde a la fecha que hubo lluvia.

```python
df_precip_clean['mes'] = df_precip_clean['fecha'].dt.month
df_precip_clean['año'] = df_precip_clean['fecha'].dt.year
df_precip_clean['estacion'] = df_precip_clean['fecha'].apply(obtener_estacion_chile)
#Verificar conversión y columnas nuevas
df_precip_clean.head()
```

Eliminar columna 'fecha'

```python
df_precip_clean.drop(['fecha'], axis=1, inplace=True)
#Verificar eliminación de columna fecha
df_precip_clean.head()
```

Definición de zonas norte, centro y sur de Chile según latitud y longitud

Conversión a float de latitud y longitud y a valores positivos para facilitar mejor aún los rangos de zonas límites

```python
df_precip_clean['latitud'] = df_precip_clean['latitud'].astype(float)*-1
df_precip_clean['longitud'] = df_precip_clean['longitud'].astype(float)*-1
```

```python
def determinar_zona(lat, lon):
    if (lat >= 17.0 and lat <= 26.0):
        return 'norte'
    elif (lat > 26.0 and lat <= 35.0):
        return 'centro'
    else:
        return 'sur'

df_precip_clean['zona'] = df_precip_clean.apply(lambda row: determinar_zona(row['latitud'], row['longitud']), axis=1)
```

```python
df_precip_clean.head()
```

Creación de variable objetivo (target) a partir de la columna 'mediciones'.

```python
def definir_inundacion(mediciones, zona):
    if mediciones >= 10 and mediciones <= 20 and zona == 'norte':
        return 1
    elif mediciones >= 50 and mediciones <= 100 and zona == 'centro':
        return 1
    elif mediciones >= 100 and mediciones <= 200 and zona == 'sur':
        return 1
    else:
        return 0

df_precip_clean['inundacion'] = df_precip_clean.apply(lambda row: definir_inundacion(row['mediciones'], row['zona']), axis=1)
```
