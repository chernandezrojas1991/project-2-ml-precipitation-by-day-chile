**Obtener mes y estación del año según fecha y eliminar columna fecha**

Crear función para determinar estación del año en Chile según fecha

```python
def obtener_estacion_chile(fecha):
    mes = fecha.month
    dia = fecha.day
    if (mes == 12 and dia >= 21) or (mes in [1, 2]) or (mes == 3 and dia <= 20):
        return 1 #verano
    elif (mes == 3 and dia >= 21) or (mes in [4, 5]) or (mes == 6 and dia <= 20):
        return 2 #otoño
    elif (mes == 6 and dia >= 21) or (mes in [7, 8]) or (mes == 9 and dia <= 20):
        return 3 #invierno
    else:
        return 4 #primavera
```

Convertir columna *fecha* en tipo datetime

```python
df_precip_clean['fecha'] = pd.to_datetime(df_precip_clean['fecha'])
```

Crear columnas *mes* y *estación* extraídas y procesadas a través de la columna *fecha*, la cual corresponde a la fecha que hubo lluvia.

```python
df_precip_clean['mes'] = df_precip_clean['fecha'].dt.month
df_precip_clean['año'] = df_precip_clean['fecha'].dt.year
df_precip_clean['estacion'] = df_precip_clean['fecha'].apply(obtener_estacion_chile)
df_precip_clean['estacion'] = df_precip_clean['estacion'].astype(int)
#Verificar conversión y columnas nuevas
df_precip_clean.head()
```

**Definición de zonas norte, centro y sur de Chile según latitud y longitud**

Conversión a float de latitud y longitud y a valores positivos para facilitar mejor aún los rangos de zonas límites

```python
df_precip_clean['latitud'] = df_precip_clean['latitud'].astype(float)*-1
df_precip_clean['longitud'] = df_precip_clean['longitud'].astype(float)*-1
```

```python
def determinar_zona(lat, lon):
    if (lat >= 17.0 and lat <= 26.0):
        return 1 #norte
    elif (lat > 26.0 and lat <= 35.0):
        return 2 #centro
    else:
        return 3 #sur

df_precip_clean['zona'] = df_precip_clean.apply(lambda row: determinar_zona(row['latitud'], row['longitud']), axis=1)
df_precip_clean['zona'] = df_precip_clean['zona'].astype(int)
```

Identificar valores atípicos para *mediciones*

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Filtrar datos por zonas
df_precip_clean_north = df_precip_clean[df_precip_clean['zona'] == 1]['mediciones']
df_precip_clean_center = df_precip_clean[df_precip_clean['zona'] == 2]['mediciones']
df_precip_clean_south = df_precip_clean[df_precip_clean['zona'] == 3]['mediciones']

# Calcular promedios por zona
mean_meditions_north = df_precip_clean_north.mean()
mean_meditions_center = df_precip_clean_center.mean()
mean_meditions_south = df_precip_clean_south.mean()

# Preparar los datos para el boxplot
data_to_plot = [df_precip_clean_north, df_precip_clean_center, df_precip_clean_south]

# Graficar el boxplot
plt.figure(figsize=(8, 6))  # Ajusta el tamaño de la figura si es necesario
plt.boxplot(data_to_plot, labels=[' Zona Norte', 'Zona Centro', 'Zona Sur'])

# Agregar puntos rojos para mostrar los promedios
plt.plot([1, 2, 3], [mean_meditions_north, mean_meditions_center, mean_meditions_south], 'ro')  # Puntos rojos

# Etiquetas y título
plt.ylabel('Milímetros de precipitaciones')
plt.title('Boxplot de Precipitaciones por Zona')

# Mostrar la gráfica
plt.show()
```

Al analizar los boxplot generados para cada una de las zonas por las cuales se divide Chile (Norte, Centro, Sur), es claro ver que existen mucho valores atípicos para cada zona. Para efectos de este proyecto, los valores atípicos no serán tratados, puesto que Chile constantemente se ve enfrentado a intensas lluvias en todas las estaciones del año, sobre todo en la zona centro y sur del país.

Creación de variable objetivo (target) a partir de la columna *mediciones* y *zona*

```python
def definir_inundacion(mediciones, zona):
    if mediciones >= 10 and mediciones <= 20 and zona == 1:
        return 1
    elif mediciones >= 50 and mediciones <= 100 and zona == 2:
        return 1
    elif mediciones >= 100 and mediciones <= 200 and zona == 3:
        return 1
    else:
        return 0

df_precip_clean['inundacion'] = df_precip_clean.apply(lambda row: definir_inundacion(row['mediciones'], row['zona']), axis=1)
```

**Análisis de los meses más lluviosos**

Como se mencionó en la interpretación del histograma, se hará un gráfico de barras para analizar el comportamiento de las precipitaciones en los meses más lluviosos del año (junio, julio, agosto) desde el año 2015 al 2020.

```python
#Filtrado por año entre 2015 y 2020, cuyos meses sean junio, julio y agosto
df_rainiest_month_year = df_precip_clean[df_precip_clean['año'].between(2015, 2020) & df_precip_clean['mes'].isin([6, 7, 8])]

#Agrupar por año y mes y obtener máximo de precipitaciones registradas
grouped = df_rainiest_month_year.groupby(['año', 'mes'])['mediciones'].max().reset_index()

#Crear tabla pivote para registrar sólo año, mes en cuestión y su medición máxima de precipitaciones (milímetros)
pivot_table = grouped.pivot(index='año', columns='mes', values='mediciones')
pivot_table.columns = ['Junio', 'Julio', 'Agosto']
```

```python
pivot_table.plot(kind='bar', figsize=(10, 6))

plt.title('Valores máximos expresado en milímetros precipitaciones por Año en los meses Junio, Julio y Agosto)')
plt.xlabel('Año')
plt.ylabel('Mediciones Máximas')
plt.legend(title='Mes')
plt.show()
```

Dado el gráfico de barras generado, se hacen las siguientes observaciones:

- En el año 2018 y 2019 se registraron los máximos valores de precipitaciones, con énfasis en el mes de julio del 2018 y el mes de junio del 2019.

- Los meses con menores valores máximos de precipitaciones son junio del 2016 y julio del 2018.

**Análisis de cantidad de inundaciones en Otoño e Invierno por año en la zona sur de Chile**

```python
#Filtrado por año entre 2015 y 2020, con inundación efectiva, estaciones del año: otoño e invierno en la zona sur de Chile
df_flood_winter_year = df_precip_clean[(df_precip_clean['año'].between(2015, 2020)) & (df_precip_clean['inundacion'] == 1) & (df_precip_clean['zona'] == 3) & (df_precip_clean['estacion'].between(2,3))]

#Agrupar por año y estacion y obtener cantidad de inundaciones registradas
grouped = df_flood_winter_year.groupby(['año','estacion'])['inundacion'].count().reset_index()

#Crear tabla pivote para registrar sólo año y cantidad de inundaciones registradas
pivot_table = grouped.pivot(index='año', columns='estacion', values='inundacion')
pivot_table.columns = ['Otoño', 'Invierno']
pivot_table.head()
```

```python
plt.figure(figsize=(6, 4))
plt.plot(pivot_table.index, pivot_table['Otoño'], marker='o', linestyle='-', color='b', label='Otoño')  # Línea para el otoño
plt.plot(pivot_table.index, pivot_table['Invierno'], marker='o', linestyle='-', color='r', label='Invierno')  # Línea para el invierno
plt.title('Cantidad de Inundaciones en Invierno (Zona Sur, Chile)')
plt.xlabel('Año')  # Etiqueta del eje x
plt.ylabel('Cantidad de veces que hubo inundación')  # Etiqueta del eje y
plt.xticks(pivot_table.index)  # Marcas en el eje x para cada año
plt.grid(True)
plt.legend()

# Mostrar el gráfico
plt.tight_layout()  # Ajustar diseño
plt.show()
```

En el gráfico de puntos es claro ver que:

- Para el año 2015 y 2016 la cantidad de inundaciones registradas en Otoño e Invierno son casi iguales.

- Se observa un incremento brusco desde el 2016 al 2017 en cuanto a las inundaciones registradas en invierno de **6 a 37** episodios, mientras que para las inundaciones registradas en otoño tuvo una leve alza de **6 a 12** episodios.

- Las inundaciones en invierno van a la baja desde el invierno de 2017 en adelante, alcanzando una cifra de **12 inundaciones**.

**Análisis de la cantidad de inundaciones en Otoño e Invierno en la zona Norte de Chile**

```python
#Filtrado por año entre 2015 y 2020, con inundación efectiva, estaciones del año: otoño e invierno en la zona norte de Chile
df_flood_winter_year = df_precip_clean[(df_precip_clean['año'].between(2015, 2020)) & (df_precip_clean['inundacion'] == 1) & (df_precip_clean['zona'] == 1) & (df_precip_clean['estacion'].between(2,3))]

#Agrupar por año y estacion y obtener cantidad de inundaciones registradas
grouped = df_flood_winter_year.groupby(['año','estacion'])['inundacion'].count().reset_index()

#Crear tabla pivote para registrar sólo año y cantidad de inundaciones registradas
pivot_table = grouped.pivot(index='año', columns='estacion', values='inundacion')
pivot_table.columns = ['Otoño', 'Invierno']
pivot_table.head()
```

```python
plt.figure(figsize=(6, 4))
plt.plot(pivot_table.index, pivot_table['Otoño'], marker='o', linestyle='-', color='b', label='Otoño')  # Línea para el otoño
plt.plot(pivot_table.index, pivot_table['Invierno'], marker='o', linestyle='-', color='r', label='Invierno')  # Línea para el invierno
plt.title('Cantidad de Inundaciones en Invierno (Zona Norte, Chile)')
plt.xlabel('Año')  # Etiqueta del eje x
plt.ylabel('Cantidad de veces que hubo inundación')  # Etiqueta del eje y
plt.xticks(pivot_table.index)  # Marcas en el eje x para cada año
plt.grid(True)
plt.legend()

# Mostrar el gráfico
plt.tight_layout()  # Ajustar diseño
plt.show()
```

En el gráfico de puntos es claro ver que:

- Las inundaciones **se mantuvieron casi constantes** durante los inviernos de los años **2015, 2016, 2017 y 2018**.

- Se muestra una **fuerte alza** de registro de inundaciones para el invierno del **2019**, alcanzando la cifra de **50 episodios**, distribuidos por toda la zona Norte de Chile.

- Los otoños de los años **2015, 2016, 2017 y 2018** presentan un zigzagueo de la presencia de inundaciones. Mientras que para el año **2019** presenta una fuerte alza de inundaciones alcanzando la cifra de **40 episodios**. Va a la baja fuertemente para el año 2020.

Nota: Comparando los dos gráficos de puntos, se puede decir que ha llovido más durante otoño del año 2019 que el invierno de 2017.
