**Obtener mes y estación del año según fecha y eliminar columna fecha**

Crear función para determinar estación del año en Chile según fecha

```python
def obtener_estacion_chile(fecha):
    mes = fecha.month
    dia = fecha.day
    if (mes == 12 and dia >= 21) or (mes in [1, 2]) or (mes == 3 and dia <= 20):
        return 1 #verano
    elif (mes == 3 and dia >= 21) or (mes in [4, 5]) or (mes == 6 and dia <= 20):
        return 2 #otoño
    elif (mes == 6 and dia >= 21) or (mes in [7, 8]) or (mes == 9 and dia <= 20):
        return 3 #invierno
    else:
        return 4 #primavera
```

Convertir columna *fecha* en tipo datetime

```python
df_precip_clean['fecha'] = pd.to_datetime(df_precip_clean['fecha'])
```

Crear columnas *mes* y *estación* extraídas y procesadas a través de la columna *fecha*, la cual corresponde a la fecha que hubo lluvia.

```python
df_precip_clean['mes'] = df_precip_clean['fecha'].dt.month
df_precip_clean['año'] = df_precip_clean['fecha'].dt.year
df_precip_clean['estacion'] = df_precip_clean['fecha'].apply(obtener_estacion_chile)
df_precip_clean['estacion'] = df_precip_clean['estacion'].astype(int)
#Verificar conversión y columnas nuevas
df_precip_clean.head()
```

Eliminar columna *fecha*

```python
df_precip_clean.drop(['fecha'], axis=1, inplace=True)
#Verificar eliminación de columna fecha
df_precip_clean.head()
```

**Definición de zonas norte, centro y sur de Chile según latitud y longitud**

Conversión a float de latitud y longitud y a valores positivos para facilitar mejor aún los rangos de zonas límites

```python
df_precip_clean['latitud'] = df_precip_clean['latitud'].astype(float)*-1
df_precip_clean['longitud'] = df_precip_clean['longitud'].astype(float)*-1
```

```python
def determinar_zona(lat, lon):
    if (lat >= 17.0 and lat <= 26.0):
        return 1 #norte
    elif (lat > 26.0 and lat <= 35.0):
        return 2 #centro
    else:
        return 3 #sur

df_precip_clean['zona'] = df_precip_clean.apply(lambda row: determinar_zona(row['latitud'], row['longitud']), axis=1)
df_precip_clean['zona'] = df_precip_clean['zona'].astype(int)
```

Identificar valores atípicos para *mediciones*

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Filtrar datos por zonas
df_precip_clean_north = df_precip_clean[df_precip_clean['zona'] == 1]['mediciones']
df_precip_clean_center = df_precip_clean[df_precip_clean['zona'] == 2]['mediciones']
df_precip_clean_south = df_precip_clean[df_precip_clean['zona'] == 3]['mediciones']

# Calcular promedios por zona
mean_meditions_north = df_precip_clean_north.mean()
mean_meditions_center = df_precip_clean_center.mean()
mean_meditions_south = df_precip_clean_south.mean()

# Preparar los datos para el boxplot
data_to_plot = [df_precip_clean_north, df_precip_clean_center, df_precip_clean_south]

# Graficar el boxplot
plt.figure(figsize=(8, 6))  # Ajusta el tamaño de la figura si es necesario
plt.boxplot(data_to_plot, labels=[' Zona Norte', 'Zona Centro', 'Zona Sur'])

# Agregar puntos rojos para mostrar los promedios
plt.plot([1, 2, 3], [mean_meditions_north, mean_meditions_center, mean_meditions_south], 'ro')  # Puntos rojos

# Etiquetas y título
plt.ylabel('Milímetros de precipitaciones')
plt.title('Boxplot de Precipitaciones por Zona')

# Mostrar la gráfica
plt.show()
```

Al analizar los boxplot generados para cada una de las zonas por las cuales se divide Chile (Norte, Centro, Sur), es claro ver que existen mucho valores atípicos para cada zona. Cabe destacar que para la zona Sur es en donde más llueve, lo cual es sabido por cultura general, datos climáticos disponibles en Internet y la literatura.

Para efectos de este proyecto, los valores atípicos no serán tratados, puesto que Chile constantemente se ve enfrentado a intensas lluvias en todas las estaciones del año, sobre todo en la zona centro y sur del país.

Creación de variable objetivo (target) a partir de la columna *mediciones* y *zona*

```python
def definir_inundacion(mediciones, zona):
    if mediciones >= 10 and mediciones <= 20 and zona == 1:
        return 1
    elif mediciones >= 50 and mediciones <= 100 and zona == 2:
        return 1
    elif mediciones >= 100 and mediciones <= 200 and zona == 3:
        return 1
    else:
        return 0

df_precip_clean['inundacion'] = df_precip_clean.apply(lambda row: definir_inundacion(row['mediciones'], row['zona']), axis=1)
```

Mapa de calor para identificar correlaciones fuertes

```python
#Crear matriz de correlación
corr_matrix = df_precip_clean.corr()
```

```python
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu')
plt.title('Mapa de calor de correlación en conjunto de datos de precipitaciones')
plt.show()
```

A modo general, no hay correlaciones muy fuertes. La estación y la zona tiene correlación del 15%, latitud con codigo_cuenca y codigo_sub_cuenca tiene estrecha relación con un 87% dado que éstas son únicas (por ejemplo, no existen dos codigo_cuenca con valor 205) y ésta se repite varias veces en una latitud/longitud porque registró miles de datos. Una correlación no tan estrecha pero sí a considerar es *mediciones* con *inundación* de un 32%. Sabemos que los milímetros caídos de precipitaciones puede estar relacionado con inundaciones, pero no todos los efectos climáticos derivados de lluvia producen inundaciones. Otra correlación un tanto fuerte es *altura* con *inundación* con un 21%. Ya que las estaciones metereológicas están instaladas en altura, la percepción de agua-lluvia es más alta que en otros puntos del país que son más bajos.
